```python
import QMMM # подключение модуля с классом QMMM (файл QMMM.py)
```

***Пример 1***

```python
file_gro = '/home/domain/data/avetrov/25_26/test_diala/npt.gro'
file_top = '/home/domain/data/avetrov/25_26/test_diala/topol.top'

file_ondx = '/home/domain/data/avetrov/25_26/test_diala/qm.ndx'
file_ogro = '/home/domain/data/avetrov/25_26/test_diala/qm.gro'
file_otop = '/home/domain/data/avetrov/25_26/test_diala/qm.top'

file_hsd = '/home/domain/data/avetrov/25_26/test_diala/dftb_in.hsd'

qm = QMMM.QM(file_gro,file_top,file_ogro,file_otop,file_ondx) #загружаем топологию и gro, обозначаем имена выходных файлов.
qm.choose_qm_to_extend('@11') # расширяем остаток цистеина взяв за основу один атом боковой цепи. В результате вся боковая цепь цистеина будет помещена в QM
qm.choose_qm_manually(f'(({qm.qm_extend_mask})<:3.0)&(:SOL)') #  берем воду в пределах 3 ангстрем вокруг боковой цепи цистеина
qm.job(qm_aim_charge=0,scheme='RC') # суммарный заряд  QM = 0, схема перераспределения заряда RC.
qm.make_hsd(file_hsd=file_hsd) # записываем файл dftb_in.hsd по QM системе.
```

***Пример 2***

```python
file_gro = '/home/domain/data/avetrov/25_26/test_PAR/2/qm-start.gro'
file_top = '/home/domain/data/avetrov/25_26/test_PAR/2/topol.top'
file_ndx = '/home/domain/data/avetrov/25_26/test_PAR/2/index.ndx'

file_ondx = '/home/domain/data/avetrov/25_26/test_PAR/2/qm.ndx'
file_ogro = '/home/domain/data/avetrov/25_26/test_PAR/2//qm.gro'
file_otop = '/home/domain/data/avetrov/25_26/test_PAR/2/qm.top'

file_hsd = '/home/domain/data/avetrov/25_26/test_PAR/2/dftb_in.hsd'

qm = QMMM.QM(file_gro,file_top,file_ogro,file_otop,file_ondx) #загружаем топологию и gro, обозначаем имена выходных файлов.
qm.choose_qm_to_extend('@2270,3347,3051') # берем в qm параоксон, боковые цепи аспартата и тирозина, выбрав по 1 атому в них 
qm.choose_qm_manually(f'(({qm.qm_extend_mask})<:5.0)&(:SOL)') # добавляем в QM воду в пределах 5 ангстрем вокруг QM
qm.job(qm_aim_charge=-1,scheme='amber') # суммарный заряд  QM = -1, схема перераспределения заряда amber.
qm.make_hsd(file_hsd=file_hsd) # записываем файл dftb_in.hsd по QM системе.
```
